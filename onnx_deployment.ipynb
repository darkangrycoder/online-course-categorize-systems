```python
df = pd.read_csv("udemy_courses_details.csv")
df.head()
```

```python
df = df.dropna().reset_index(drop=True)
df.shape
```

```python
import collections
from collections import Counter
df = df.dropna().reset_index(drop=True)
genre_list = []  # List to store genres
category_count = Counter()
indices_to_drop = []  # Define indices_to_drop here

for idx, labels_str in enumerate(df['Labels']):
    categories_str = labels_str.strip("[]").replace("'", "")  # Remove brackets and single quotes
    categories = categories_str.split(',')  # Split by comma to create a list
    if len(categories):
        genre_list.append(categories[0])
    else:
        indices_to_drop.append(idx)

    # Process categories for counting
    category_count.update([cat.strip() for cat in categories])

# Drop the rows based on indices_to_drop and reset index
df = df.drop(indices_to_drop).reset_index(drop=True)

print(f"Number of Genres: {len(genre_list)}")
print(genre_list)
print(f"Number of Categories: {len(category_count)}")
print(category_count)
```

```python
threshold = int(len(df) * 0.01)
rare_categories = [key for key, value in category_count.items() if value <= threshold]
print(f"Number of Rare Categories: {len(rare_categories)}")
print(rare_categories[:50])
```

```python

revised_genre_list = []
indices_to_drop = []

for idx, labels_str in enumerate(df['Labels']):
    genre_list = labels_str.split(',')
    revised_genre = [label for label in genre_list if label.strip() not in rare_categories]

    if len(revised_genre) == 0:
        indices_to_drop.append(idx)
    else:
        revised_genre_list.append(revised_genre)

# Drop rows with no valid genres and reset index
df = df.drop(indices_to_drop).reset_index(drop=True)
print(df.shape)
```

```python
df['revised_genres'] = revised_genre_list
```

```python
revised_genres_list = df.revised_genres.to_list()
revised_genre_count = {}

for genre_list in revised_genres_list:
    for label in genre_list:
        if label in revised_genre_count.keys():
            revised_genre_count[label] += 1
        else:
            revised_genre_count[label] = 1

print(f"Number of genres: {len(revised_genre_count)}")
print(revised_genre_count)
```

```python
encode_genre_types = {key: idx for idx, (key, value) in enumerate(revised_genre_count.items())}
with open("genre_types_encoded.json", "w") as fp:
  json.dump(encode_genre_types, fp)
```

```python
categorial_genre_list = []
revised_genres_list = df.revised_genres.to_list()

for revised_genres in revised_genres_list:
    categorial_list = [0] * len(encode_genre_types)
    for Labels in revised_genres:
        genre_type_index = encode_genre_types[Labels]
        categorial_list[genre_type_index] = 1  # Set the value to 1 to indicate presence
    categorial_genre_list.append(categorial_list)
categorial_genre_list[3][:10]
```

```python
from sklearn import metrics

def metric_measures(test_df, preds):

  targets = [np.asarray(target) for target in test_df['genre_cat_list'].to_list()]
  outputs = [np.asarray(pred) for pred in preds]


  accuracy = metrics.accuracy_score(targets, outputs)
  f1_score_micro = metrics.f1_score(targets, outputs, average='micro')
  f1_score_macro = metrics.f1_score(targets, outputs, average='macro')

  print(f"F1 Score (Micro) = {f1_score_micro}")
  print(f"F1 Score (Macro) = {f1_score_macro}")

  return
```
